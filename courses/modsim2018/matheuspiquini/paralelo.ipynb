```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Lslack = 0.223
Umax = 0.04
Lceopt = 0.093
width = 0.63*Lceopt
Fmax = 7400
a = 0.25*Fmax
b = 0.25*10*Lceopt

Lce = 0.087
h = 0.001
t = 0
tend = 2.99
i = 1

data = np.zeros((int(tend/h)+3, 2))
```

```python
while t < tend:
    if t <= 1:
        Lm = 0.31
    if t > 1 and t < 2:
        Lm = 0.31 - 0.04*(t-1)
    Lsee = Lm - Lce
    if Lsee < Lslack:
        Fsee = 0
    else:
        Fsee = Fmax*((Lsee-Lslack)/(Umax*Lslack))**2
    if Lce < Lceopt:
        Fkp = 0
    else:
        Fkp = Fmax*((Lsee-Lceopt)/*(Umax*Lceopt))**2
    F0 = max(0, (Fmax*(1-((Lce-Lceopt)/width)**2))
    if (Fsee - Fkp) > F0:
        print ('error')
    Lcedot = -b*(F0-(Fsee-Fkp))/(Fsee+a-Fkp)
    Lce = Lce + h*Lcedot
    t = t + h
    i = i + 1
    data[i,0] = t
    data[i,1] = Fsee - Fkp

plt.plot(data[:,0], data[:,1])
```
